```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Datasets/ANZ.csv", index_col=['extraction'], parse_dates=['extraction'])

df1 = pd.DataFrame()

dayavg1 = df.resample('1D').amount.mean()

dayavg1.plot()
plt.xlabel('date')
plt.ylabel('amount')
plt.show()

dayavg2 = df.resample('2D').amount.mean()
print (dayavg2)

dayavg2.plot()
plt.xlabel('date')
plt.ylabel('amount')
```

```python
dayavg7 = df.resample('7D').amount.mean()
print(dayavg7)

df2 = pd.DataFrame()

df2 = dayavg7

df2.plot()
plt.xlabel('date')
plt.ylabel('amount')
```

```python
plt.boxplot(df2,vert=False)
plt.yticks([1], ['7 Day Avg.'])
plt.xlabel("Amount")
```
